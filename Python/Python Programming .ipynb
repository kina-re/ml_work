```python
def matrix_mul(A, B):
    """
    Get the product of two matrices
    :param A: list of list containing numbers
    :param B: list of list containing numbers
    :return: product of two matrices as list of list containing numbers
    """
    # if any of the matrices is empty
    if len(A) == 0 or len(B) == 0:
        print('Matrix should not be empty')

    # when no.of cols in 1st matrix equal to no. of rows in 2nd matrix
    try:

        # create ans matrix with zeroes equal to cols in second matrix and rows in first matrix
        _: int 
        ans = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]

        # iterate through the rows of matrix A
        for i in range(len(A)):

            # iterate through the cols of matrix B
            for j in range(len(B[0])):

                # iterate through the rows of matrix B
                for k in range(len(B)):
                    ans[i][j] += A[i][k] * B[k][j]
        return ans

    except IndexError:
        print("No.of cols in 1st matrix must be equal to no. of rows in 2nd matrix")
```

```python
# https://stackoverflow.com/questions/16489449/select-element-from-array-with-probability-proportional-to-its-value
from random import uniform


def pick_a_number_from_list(A):
    """
    selects an element from the list A randomly with probability proportional to its magnitude. 
     does the same experiment for 100 times with replacement, in each experiment prints a number 
     that is selected randomly from A.

    :param A: list of integers
    :return: dictionary with key value pair of number and its frequency
    """
    # if the list is empty
    if len(A) == 0:
        print('List should not be empty')

    # total all the values of list
    addition = sum(int(i) for i in A)

    # find proportion for all values in list
    prop_list = [int(i) / addition for i in A]
    total = 0

    # total of all proportional values
    cum_sum = []

    for j in range(len(prop_list)):
        total += prop_list[j]
        cum_sum.append(total)

    # random uniform value between 0 and 1
    r = uniform(0, 1)

    # if the random value is less than the value in cumulative sum of proportional values then return that cumsum value
    for i, k in enumerate(cum_sum):
        if r <= k:
            return i


def sampling_based_on_magnitude():
    sampl = {}

    for i in range(1, 100):

        # get the index value of a number which is greater than random value
        number = pick_a_number_from_list(A)

        # get the value based on index
        n = A[number]

        # if value not in dictionary add it
        if n not in sampl:
            sampl[n] = 1

        # if value is in dictionary , increment it by 1
        else:
            sampl[n] += 1
    print(sampl)


A = [0, 5, 27, 6, 13, 28, 100, 45, 10, 79]

pick_a_number_from_list(A)
sampling_based_on_magnitude()


```

<h3> Q3: Replace the digits in the string with #</h3>

consider a string that will have digits in that, we need to remove all the not digits and replace the digits with #
<pre>
Ex 1: A = 234                Output: ###
Ex 2: A = a2b3c4             Output: ###
Ex 3: A = abc                Output:   (empty string)
Ex 5: A = #2a$#b%c%561#      Output: ####
</pre>

```python
import re


def replace_digits(String):
    """
    consider a string that will have digits in that,
    remove all the not digits and replace the digits with #

    :param String:string containing alphanumeric characters and symbols
    :return: string containing #s equals to number of digits in string
    """

    if len(String) == 0:
        print('Please provide the valid string')

    # Matches any character which is not a decimal digit
    pattern = re.compile('\D')

    # replace every occurrence of pattern with ''
    no_char = pattern.sub('', String)

    # Empty string to store result
    new_str = ''

    # replace numbers with # and add to empty string
    for _ in no_char:
        new_str += '#'
    return new_str


replace_digits('#2a$#b%c%561#')
```

for a given string with comma seprate values, which will have both missing values numbers like ex: "_, _, x, _, _, _"

```python
def curve_smoothing(string):
    """
    Filling missing values
    :param string: string of numbers and blank spaces
    :return: string of numbers with avg of adjacent numbers
    
    """

    str_modified = string.split(',')

    # replace empty spaces with 0
    str_modified = [0 if x == '_' else x for x in str_modified]

    lst = list(map(int, str_modified))

    i = 0
    # when last and first elements of list are not zero(blank)
    if lst[0] != 0 and lst[-1] != 0:
        a = -1

    # when first element is not zero(empty) and last element is zero(blank)
    elif lst[0] != 0 and lst[-1] == 0:
        lst = lst[::-1]
        a = 0
    else:
        a = 0
    # when in between elements are zero(blank)
    while i < len(lst):

        # assign initial value of i , so that it can be used in else statement
        m = i

        # when element is nonzero element
        if lst[i] != 0:

            # assign the index to k
            k = i

            # calculate the value from adjacent values
            v = (lst[a] + lst[k]) / (k - a + 1)

            # assign calculated value to list
            for n in range(a, k + 1):
                lst[n] = v
            # assign latest k to a
            a = k
        # increment the index
        i += 1
    else:
        # when last element is zero
        if lst[-1] == 0:

            # calculates the value from adjacent values
            v = (lst[a] + lst[-1]) / (m - k + 1)

            # assign value to  the new list with modified values
            for n in range(a, m + 1):
                lst[n] = v
    return ','.join(map(str, lst))


s = "_,_,30,_,_,_,50,_,_"
curve_smoothing(s)
```

e. P(F=F5|S==S1)=1/4, P(F=F5|S==S2)=0/3, P(F=F5|S==S3)=0/3
</pre>




```python
import re


def compute_conditional_probablities(A):
    """
    Computes conditional probabilities
    :param A: column of five unique values F1,F2,F3,F4,F5
    :return: conditional probabilities 
    """

    # create a list of unique f elements
    set_f = list(set([i[0] for i in A]))

    # create a list of unique s elements
    set_s = list(set([i[1] for i in A]))

    # create dictionary with s1, s2, s3 as keys and 0 as values
    dict_s = dict((el, 0) for el in set_s)

    # create combination of f elements and s elements
    comb_fs = []
    for i in set_f:
        for j in set_s:
            comb_fs.append(i + j)

    # combine elements from nested list
    comb_ele_a = [x + y for x, y in A]

    # combination of F and S as keys and 0 as values
    initial_dict = dict((el, 0) for el in comb_fs)

    for i in range(len(A)):
        if A[i][1] in dict_s:
            dict_s[A[i][1]] += 1

    for i in range(len(comb_ele_a)):
        if comb_ele_a[i] in initial_dict:
            initial_dict[comb_ele_a[i]] += 1
            
    for k,v in initial_dict.items():
        c = k[0:2]
        w = k[2:]
        a = int(v)/int(dict_s[w])
        ans = round(a,2)
     
        print('P(F={}/S=={})={}'.format(c,w,ans) )
    

   


A = [['F1', 'S1'], ['F2', 'S2'], ['F3', 'S3'], ['F1', 'S2'], ['F2', 'S3'], ['F3', 'S2'], ['F2', 'S1'], ['F4', 'S1'],
     ['F4', 'S3'], ['F5', 'S1']]

compute_conditional_probablities(A)
```

<h3> Q9: Given two sentances S1, S2</h3>
You will be given two sentances S1, S2 your task is to find 
<pre>

b. Words in S1 but not in S2
c. Words in S2 but not in S1
</pre>

Ex: 
<pre>
S1= "the first column F will contain only 5 uniques values"
S2= "the second column S will contain only 3 uniques values"
Output:
a. 7
b. ['first','F','5']
c. ['second','S','3']

```python
def string_features(S1, S2):
    """
    given two sentances S1, S2 it finds
    a. Number of common words between S1, S2
    b. Words in S1 but not in S2
    c. Words in S2 but not in S1
    :param S1: string 1
    :param S2: string 2
    :return: integer, lists
    """
    s1 = S1.split()
    s2 = S2.split()
    count = 0
    m = []
    n = []
    for w in s1:
        if w in s2:
            count += 1

        if w not in s2:
            m.append(w)
    for w in s2:
        if w not in s1:
            n.append(w)
    return count, m, n


S1 = "the first column F will contain only 5 unique values"
S2 = "the second column S will contain only 3 unique values"
string_features(S1,S2)
```

<h3> Q10: Given two sentances S1, S2</h3>
You will be given a list of lists, each sublist will be of length 2 i.e. [[x,y],[p,q],[l,m]..[r,s]]
consider its like a martrix of n rows and two columns

a. the first column Y will contain interger values <br>
b. the second column $Y_{score}$ will be having float values <br>
Your task is to find the value of $f(Y,Y_{score}) = -1*\frac{1}{n}\Sigma_{for each Y,Y_{score} pair}(Ylog10(Y_{score})+(1-Y)log10(1-Y_{score}))$ here n is the number of rows in the matrix
<pre>
Ex:
[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
output:
0.4243099
</pre>
$\frac{-1}{8}\cdot((1\cdot log_{10}(0.4)+0\cdot log_{10}(0.6))+(0\cdot log_{10}(0.5)+1\cdot log_{10}(0.5)) + ... + (1\cdot log_{10}(0.8)+0\cdot log_{10}(0.2)) )$

```python
import math


def compute_log_loss(A):
    """
    Computes log loss
    :param A: nested list of classes and probabilities associated with classes
    :return: log loss
    """

    # Get probability values associated with class 1
    v1 = [x[1] for x in A if x[0] == 1]

    # Get probability values associated with class 0
    v0 = [x[1] for x in A if x[0] == 0]

    # deducting the probability value of class zero from 1
    v0_1 = [1 - x for x in v0]

    # probability values for belonging point to class 1
    vals = v1 + v0_1

    # log of probability values
    log_vals = [math.log10(x) for x in vals]

    # sum of log values
    tot_vals = math.fsum(log_vals)

    # calculation of logloss
    log_loss = -1*(tot_vals / len(A))
    return round(log_loss, 2)


A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
loss = compute_log_loss(A)
print(loss)
```
